## Візуалізації розподілу категоріальних даних

### boxplot 
медіанне значення, квартілі та викиди

```python
import seaborn as sns
data = sns.load_dataset("mpg")
data
```

```python
sns.boxplot(x="origin", y="mpg", hue="cylinders", data=data)
```

### violinplot 
діаграмa з оцінкою щільності ядра

```python
sns.violinplot(x="origin", y="mpg", hue="origin", data=data)
```

### boxenplot 
діаграмa з прямокутників для візуалізації великих наборів даних.

```python
sns.boxenplot(x="origin", y="mpg", hue="origin", data=data)
```
