## Функція scatterplot​

Побудуємо точковий графік залежності дальності пробігу (displacement) автомобіля від витрат палива (mpg):

```python
import seaborn as sns
```

```python
data = sns.load_dataset("mpg")
sns.scatterplot(x="mpg", y="displacement", data=data)
```

```python
sns.scatterplot(x='mpg', y='displacement', hue='origin', data=data)
```
