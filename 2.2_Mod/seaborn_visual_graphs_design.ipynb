## Візуальне оформлення графіків​

### Стилі

```python
import seaborn as sns
data = sns.load_dataset("mpg")
data
```

```python
# set_style(style=None, rc=None)
```

```python
sns.set_style("darkgrid")
sns.lineplot(x="model_year", y="horsepower", hue="origin", data=data)
sns.axes_style() # список параметрів, що відповідають за оформлення графіка
```

```python
# перевизначати параметри оформлення:
sns.set_style("darkgrid",  {'axes.labelcolor':"(0.5,0.5,0)", 'axes.edgecolor':'#061358',
'xtick.color':'#0A5806'})
sns.lineplot(x='model_year', y='horsepower', hue="origin", data=data)
```

### Контекст

```python
# Для встановлення контексту:
# set_context(context=None, font_scale=1, rc=None)
```

```python
sns.set_context("poster")
sns.scatterplot(x='mpg', y='displacement', data=data) # Для отримання списку параметрів контексту
```
